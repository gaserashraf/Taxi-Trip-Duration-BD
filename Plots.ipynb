```python
import matplotlib.pyplot as plt

models = ['Linear Regression', 'Decision Tree Regression', 'Random Forest Regression', 'XGBoost Regression', 'No Max-Depth', 'StackEnsemble']
rmse_scores = [414.42, 375.42, 385.52, 337.25, 316.57, 1143.8]
r2_scores = [0.601, 0.601, 0.587, 0.736, 0.79, 0.88025]

# Plotting RMSE scores
plt.figure(figsize=(5, 5))
plt.plot(models, rmse_scores, marker='o')
plt.xlabel('Models')
plt.ylabel('RMSE')
plt.title('RMSE Scores for Different Models')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plotting R-squared scores
plt.figure(figsize=(5, 5))
plt.plot(models, r2_scores, marker='o')
plt.xlabel('Models')
plt.ylabel('R-Squared Score')
plt.title('R-Squared Scores for Different Models')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
